```python


X = np.arange(-math.pi, (math.pi + 0.05), 0.05)

cos_x = np.cos(X)

plt.plot(X, cos_x)
plt.show()
```

```python
X = np.arange(-math.pi, (math.pi + 0.05), 0.05)

cos_x = np.cos(X)
sin_x = np.sin(X)

plt.plot(X, cos_x, label='cosine x')
plt.plot(X, sin_x, label='sine x')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Sin(x) vs Cos(x)")

plt.legend(loc='upper left')

plt.show()
```

```python
from sklearn import datasets
import pandas as pd

# Load some data
iris = datasets.load_iris()         # iris is a sklearn.utils.Bunch

iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
# the df has 4 columns : sepal length (cm)  sepal width (cm)  petal length (cm)  petal width (cm)

iris_df['species'] = iris['target']
# add a 5th field 'species' which has values 0, 1 and 2 which corresponds to below...

colours = ['red', 'orange', 'blue']
species = ['I. setosa', 'I. versicolor', 'I. virginica']


# call plt.scatter()  3 times for each species x=sepal length and y = petal length

for i in range(0, 3):    
    species_df = iris_df[iris_df['species'] == i]    
    plt.scatter(        
        species_df['sepal length (cm)'],        
        species_df['petal length (cm)'],
        color=colours[i],        
        alpha=0.4,        
        label=species[i]   
    )

plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.title('Iris dataset: petal length vs sepal length')
plt.legend(loc='lower right')

plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
sepal_lengths=features[0]
sepal_widths=features[1]
petal_widths=features[3]
species=iris.target
#  the colors correspond to 1 or 3 species

plt.scatter(sepal_lengths, sepal_widths, alpha=0.2,
      s=100*petal_widths, c=species, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
print(iris.feature_names[0])
```

stacked plots

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```
