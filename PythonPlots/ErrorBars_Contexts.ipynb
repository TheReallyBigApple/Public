```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import numpy as np 
```

<center><strong>Error Bars</strong></center>
<br><br>
Error Bar, SaveFig. Note this is a <i>asymmetrical</i> error bar as e1 (above the line) and e2 (below the line) are not equal.
<br>

```python
x = np.arange(0, 4, 0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x, y, yerr=e1, fmt='.-')
plt.savefig('7900_03_09.png')
plt.clf()
e2 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x, y, yerr=e1, xerr=e2, fmt='.-', capsize=0)
plt.savefig('7900_03_10.png')
plt.clf()
plt.errorbar(x, y, yerr=[e1,e2], fmt='.-')
plt.show()
```

you can format error bars much like you can format lines

```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='*c');  # fmt * = start    c= cyan
# plt.errorbar(x, y, yerr=dy, fmt='.k');  # fmt . = dot   k = black
# plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0);
```

colorbars

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
        
```

colors

```python

# note the use of sizes, and the colorbar !

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
                   cmap='viridis')
plt.colorbar();  # show color scale
```

contexts

```python

        
def hist_and_lines():
   np.random.seed(0)
   fig, ax = plt.subplots(1, 2, figsize=(11, 4))
   ax[0].hist(np.random.randn(1000))
   for i in range(3):
       ax[1].plot(np.random.rand(10))
   ax[1].legend(['a', 'b', 'c'], loc='lower left')
   
           
hist_and_lines()

           
with plt.style.context('ggplot'):
            hist_and_lines()
            
            
with plt.style.context('fivethirtyeight'):
            hist_and_lines()


with plt.style.context('bmh'):
            hist_and_lines() 
            
with plt.style.context('dark_background'):
            hist_and_lines()            

with plt.style.context('grayscale'):
            hist_and_lines()            
            
            
```
