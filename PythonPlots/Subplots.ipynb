```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import numpy as np 
import datetime as dt 
```

```python
N_points = 100000
n_bins = 20

np.random.seed(10)

# Generate a normal distribution, center at x=0 and y=5
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5

# fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
fig, axs = plt.subplots(nrows=1, ncols=2, sharey=True, tight_layout=True)
# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
plt.show()
```

Two columns. One row. Uses subplots() to return an array of axes

```python
np.random.seed(0)
fig, ax = plt.subplots(1, 2, figsize=(11, 4))
ax[0].hist(np.random.randn(1000))
for i in range(3):
   ax[1].plot(np.random.rand(10))
ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
np.random.seed(0)
fig, ax = plt.subplots(2, 1, figsize=(4, 11))
ax[0].hist(np.random.randn(1000))
for i in range(3):
   ax[1].plot(np.random.rand(10))
ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

One column. Three rows. Uses add_subplot()

```python
mpl.rcParams['font.size'] = 4.

x = np.arange(11)           # 0,1,2,3...10
fig = plt.figure()

ax1 = fig.add_subplot(311)
ax1.set_title("Hooray")
ax1.plot(x, x)

ax2 = fig.add_subplot(312, sharex=ax1)
ax2.set_title("For")
ax2.plot(2*x, 2*x)

ax3 = fig.add_subplot(313, sharex=ax1)
ax3.set_title("Us")
ax3.plot(3*x, 3*x)
plt.show()
```

One column. Three rows. Uses subplots() to return an array of axes. Set sharex and y to True.

```python
fig, ax = plt.subplots(3, 1, figsize=(9, 9), sharex=True, sharey=True)
plt.suptitle('Different curves', fontsize=19, fontweight='bold', color='purple')
ax[0].set_title("Hooray", fontsize=15)
ax[0].plot(x,x)
ax[1].set_title("For", fontsize=15)
ax[1].plot(2*x, 2*x)
ax[2].set_title("Us", fontsize=15)
ax[2].plot(3*x, 3*x)
plt.show()
```
