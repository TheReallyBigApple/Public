The CSS4 (Cascading Style Sheet) colors are the same as the X11 specification.

See [here](https://en.wikipedia.org/wiki/X11_color_names) to get more colors.

```python

# Define The figure layout
fig = plt.figure(figsize=(12,6))
ax1 = plt.subplot(111)

# Set up data for x co-ordinates, 20 numbers equally spaced between 0 and 10
x = np.linspace(0, 10, 20)

# Demonstration of many ways of color specification
# 'xkcd:sky blue'      # name from xkcd color survey
# 'green'              # CSS4 color name
# '#1F1F1F1F'          # hexadecimal value in RGBA format
# 'b'                  # CSS4 color abbreviation
# '#1C0B2D'            # hexadecimal value in RGB format, digits range from 0 to F
# 'pink'               # tableau color
# 'C4'                 # color from property cycle, this is case sensitive, C has to be capital

# color_list = ['blue violet', 'green', 'crimson', 'deep pink', 'green', 'pink', 'Turquoise']
color_list = ['powderblue','green', 'crimson','green', 'pink', 'Turquoise', 'seagreen']
for i, color in enumerate(color_list):
    y = x - (-5*i + 15)
    ax1.plot(x, y, color)
ax1.set_title('colors demo')
plt.show()
```

The codes do not overlap. You can combine them as one parameter.

```python
y = np.arange(1,3,0.3)
plt.plot(y, 'cx--', y+1, 'mo:', y+2, 'kp-.')
plt.show()
```

of course its better to be explicit

```python
y = np.arange(1,3,0.3)
plt.plot(y, linestyle ='--', color='c', marker='x')
plt.plot(y+1, linestyle =':', color='m', marker='o')
plt.plot(y+2, linestyle ='-.', color='k', marker='p')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1,3,0.3)
plt.plot(y, color='blue', linestyle='dashdot', linewidth=4,
         marker='o', markerfacecolor='red', markeredgecolor='black',
         markeredgewidth=3, markersize=12)
plt.show()
```

Override the ticks ...

```python
x=[5, 3, 7, 2, 4, 1]
plt.plot(x)
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f'])
plt.yticks(range(1, 8, 2))
plt.show()
```

Rotate the x labels

```python
x = ['Nuclear', 'Hydro', 'Gas', 'Oil', 'Coal', 'Biofuel']
energy = [5, 6, 15, 22, 24, 8]

plt.bar(x, energy, color='green')
plt.xlabel("Energy Source")
plt.ylabel("Energy Output (GJ)")
plt.title("Energy output from various fuel sources")

plt.xticks(rotation=45)
plt.show()
```

when ax3 shares axis with ax1, ax1 increases to match ax3

```python

import matplotlib as mpl
mpl.rcParams['font.size'] = 4.

x = np.arange(11)           # 0,1,2,3...10
fig = plt.figure()

ax1 = fig.add_subplot(311)
ax1.set_title("Hooray")
ax1.plot(x, x)

# ax1.get_xticks()   # array([-2.,  0.,  2.,  4.,  6.,  8., 10., 12.])


ax2 = fig.add_subplot(312, sharex=ax1)
ax2.set_title("For")
ax2.plot(2*x, 2*x)

ax3 = fig.add_subplot(313, sharex=ax1)
ax3.set_title("Us")
ax3.plot(3*x, 3*x)
plt.show()

# ax1 is of type  mpl.axes._subplots.AxesSubplot


# ax1.get_xscale()       # linear
# ax1.get_xticks()      # array([-5.,  0.,  5., 10., 15., 20., 25., 30., 35.])
```

ax1, ax2, ax3 are *subplots* but this is a derived class of Axes
[see here](https://matplotlib.org/stable/api/axes_api.html) to read about the Axes object

Just a sample of useful functions :

<pre>
    Axes.invert_yaxis		Invert the y-axis
    Axes.yaxis_inverted		Return if yaxis is in "inverse" direction
    Axes.set_xlim			Set the x-axis view limits
    Axes.get_xlim			Return the x-axis view limits
    Axes.set_xbound			Set the lower and upper numerical bounds of the x-axis
    Axes.set_xlabel			Set the label for the x-axis
    Axes.get_xlabel			Get the xlabel text string
    Axes.set_title			Set a title for the Axes
    Axes.get_title			Get an Axes title
    Axes.set_xscale			Set the x-axis scale
    Axes.get_xscale			Return the xaxis' scale (as a str)
    Axes.autoscale			Autoscale the axis view to the data (toggle)
    Axes.set_aspect			Set the aspect ratio of the axes scaling, i.e. y/x-scale
    Axes.set_xticks			Set the xaxis' tick locations and optionally labels
    Axes.set_xticklabels	Set the xaxis' labels with list of string labels
    Axes.legend			    Place a legend on the Axes
</pre>
