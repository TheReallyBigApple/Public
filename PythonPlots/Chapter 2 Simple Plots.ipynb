```python
import matplotlib.pyplot as plt
plt.plot([1, 3, 2, 4,3,5,4,6])
plt.show()
```

<br><br>
Python differs from R by starting its indexes at 0, not 1
<br><br>

```python
x = range(6) 
plt.plot(x, [x**2 for x in x])
plt.show()
```

use arange to create a non-integer range
note: without calling show(), we have some extra output

```python
import numpy as np 
x = np.arange(0.0, 6.0, 0.01) 
plt.plot(x, [x**2 for x in x]) 
```

If python is running interactively, then you dont need show()

```python
plt.isinteractive()
```

You can turn interactivity on and off with plt.ion and plt.ioff

You can call plot several times. Colors automatically change.

```python
x = range(1, 5) 

plt.plot(x, [1.5*xi for xi in x])
plt.plot(x, [3.0*xi for xi in x])
plt.plot(x, [xi/3.0 for xi in x])
plt.show()
```

To add a grid, call plt.grid(True)
To adjust the ticks, call plt.axis()

```python
x = np.arange(1, 5)
plt.plot(x, x*1.5, x, x*3.0, x, x/3.0)
plt.axis([0, 5, -1, 13])
plt.grid(True)
plt.show()
```

subplots_adjust squeezes the shape of the plot. Left/Right or Top/Bottom look the same.

```python
plt.plot([1, 3, 2, 4])
plt.xlabel('This is the X axis')
plt.ylabel('This is the Y axis')
plt.subplots_adjust(bottom=0.53)
plt.show()
```

```python
plt.plot([1, 3, 2, 4])
plt.xlabel('This is the X axis')
plt.ylabel('This is the Y axis')
plt.subplots_adjust(top=0.53)
plt.show()
```

```python
plt.plot([1, 3, 2, 4])
plt.xlabel('This is the X axis')
plt.ylabel('This is the Y axis')
plt.subplots_adjust(left=0.53)
plt.show()
```

```python
plt.plot([1, 3, 2, 4])
plt.xlabel('This is the X axis')
plt.ylabel('This is the Y axis')
plt.subplots_adjust(right=0.53)
plt.show()
```

Use of title() and legend(). Note how legend uses the labels.

```python
x = np.arange(1, 5)
plt.plot(x, x*1.5, label='Normal')
plt.plot(x, x*3.0, label='Fast')
plt.plot(x, x/3.0, label='Slow')
plt.legend()
plt.title('Simple plot')
plt.show()
```

Now bring everything together. Note the loc ( or location parameter ) for legend()

```python
x = np.arange(1, 5)
plt.plot(x, x*1.5, label='Normal')
plt.plot(x, x*3.0, label='Fast')
plt.plot(x, x/3.0, label='Slow')
plt.grid(True)
plt.title('Sample Growth of a Measure')
plt.xlabel('Samples')
plt.ylabel('Values Measured')
plt.legend(loc='upper left')
plt.subplots_adjust(bottom=0.13)
plt.show()
```

you can position loc manually

<br><br>
loc=(-0.1, 0.9) puts the legend at the bottom left below the x-axis
